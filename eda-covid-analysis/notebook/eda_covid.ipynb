## Importação e carregamento

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregamento do arquivo
df = pd.read_csv('../data/covid_data.csv')

# Visualizar as 5 primeiras linhas
df.head()
```

### Verificar informações básicas

```python
# Verificar colunas, tipos e valores nulos
df.info()
```

### Estatísticas descritivas

```python
# Estatísticas numéricas
df.describe()
```

### Verificar valores ausentes

```python
# Quantidade de nulos por coluna
df.isnull().sum().sort_values(ascending=False)
```

### Limpeza de dados - Fazer uma análise simples

```python
df_filtrado = df[['country', 'date', 'total_cases', 'total_deaths', 'population']]
```

```python
# Converter data
df_filtrado['date'] = pd.to_datetime(df_filtrado['date'])
```

```python
# Verificando se tem dados nulos nas colunas
df_filtrado.isnull().any()
```

```python
# Vamos remover linhas com nulos em colunas críticas
df_filtrado = df_filtrado.dropna(subset=['total_cases', 'total_deaths'])
```

```python
# Vamos criar uma coluna adicional que representa a Taxa de mortalidade (mortes / casos)
import numpy as np

df_filtrado['death_rate'] = np.where(
    df_filtrado['total_cases'] == 0,
    0.0,
    df_filtrado['total_deaths'] / df_filtrado['total_cases']
)
```

### Exploração Inicial

```python
# Os 10 países com mais mortes acumuladas

# Ordena os dados pela coluna date (do mais antigo para o mais recente)
# Remove linhas duplicadas de acordo com a coluna country, mantendo a última ocorrência (ou seja, a com a data mais recente).
df_latest = df_filtrado.sort_values('date').drop_duplicates('country', keep='last')

# Seleciona apenas as colunas country e total_deaths.
# Ordena os países de forma decrescente pelo total de mortes acumuladas
# Pega os 10 primeiros (ou seja, os 10 países com mais mortes).
df_latest[['country', 'total_deaths']].sort_values(by='total_deaths', ascending=False).head(10)
```

### Visualizações iniciais

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Evolução das mortes no Brasil
brasil = df_filtrado[df_filtrado['country'] == 'Brazil']

plt.figure(figsize=(10, 5))
sns.lineplot(data=brasil, x='date', y='total_deaths')
plt.title('Total de Mortes por COVID-19 no Brasil')
plt.ylabel('Mortes acumuladas')
plt.xlabel('Data')
plt.grid(True)
plt.tight_layout

# Salvar imagem na pasta /images
plt.savefig('../images/total_mortes_no_brasil.png')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Lista de países para comparar
paises = ['Brazil', 'United States', 'India', 'United Kingdom', 'Italy']

# Filtrar dados para os países de interesse
df_comparativo = df_filtrado[df_filtrado['country'].isin(paises)]

# Plotar
plt.figure(figsize=(14, 6))
sns.lineplot(data=df_comparativo, x='date', y='total_deaths', hue='country')

plt.title('Comparativo de Mortes por COVID-19 - Brasil e Outros Países')
plt.xlabel('Data')
plt.ylabel('Total de Mortes Acumuladas')
plt.legend(title='País')
plt.grid(True)
plt.tight_layout()

# Salvar imagem na pasta /images
# plt.savefig('../images/comparativo_mortes_brasil_outros_paises.png')

plt.show()
```

### Gráfico: Mortes diárias por COVID-19 no Brasil

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar apenas dados do Brasil
brasil = df_filtrado[df_filtrado['country'] == 'Brazil'].copy()

# Garantir que está ordenado por data
brasil = brasil.sort_values('date')

# Calcular mortes diárias
brasil['daily_deaths'] = brasil['total_deaths'].diff().fillna(0)

# Corrigir valores negativos (casos de ajustes nos dados)
brasil['daily_deaths'] = brasil['daily_deaths'].apply(lambda x: x if x >= 0 else 0)

# Plotar gráfico
plt.figure(figsize=(14, 6))
sns.lineplot(data=brasil, x='date', y='daily_deaths', color='red')

plt.title('Mortes Diárias por COVID-19 no Brasil')
plt.xlabel('Data')
plt.ylabel('Mortes por dia')
plt.grid(True)
plt.tight_layout()

# Salvar imagem na pasta /images
# plt.savefig('../images/mortes_diarias_brasil.png')

plt.show()
```

### Taxa de Mortalidade por País (Casos vs Mortes)

```python
# Último registro por país
df_ultimos = df_filtrado.sort_values('date').drop_duplicates('country', keep='last')

# Filtrar países com mais de 100.000 casos para evitar distorções
df_taxa = df_ultimos[df_ultimos['total_cases'] > 100000]

# Top 10 países com maior taxa de mortalidade
df_taxa_top10 = df_taxa.sort_values('death_rate', ascending=False).head(10)
```

```python
# Grafico
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.barplot(data=df_taxa_top10, x='death_rate', y='country', palette='Reds_r', hue='country', legend=False)

plt.title('Top 10 Países com Maior Taxa de Mortalidade por COVID-19')
plt.xlabel('Taxa de Mortalidade (mortes / casos)')
plt.ylabel('País')
plt.tight_layout()

plt.savefig('../images/taxa_mortalidade_top10.png')

plt.show()
```

### Mortes por 100 Mil habitantes

```python
df_filtrado['deaths_per_100k'] = (df_filtrado['total_deaths'] / df_filtrado['population']) * 100000

df_mortalidade = df_filtrado.sort_values('date').drop_duplicates('country', keep='last')
df_mortalidade = df_mortalidade[df_mortalidade['total_deaths'] > 1000]  # Filtro opcional

top10_100k = df_mortalidade.sort_values('deaths_per_100k', ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=top10_100k, x='deaths_per_100k', y='country', hue='country', legend=False, palette='Oranges_r')

plt.title('Top 10 Países com Mais Mortes por 100 Mil Habitantes')
plt.xlabel('Mortes por 100k habitantes')
plt.ylabel('País')
plt.tight_layout()

plt.savefig('../images/mortes_100k_top10.png')

plt.show()
```

### Heatmap de Mortes Diárias por Mês e Ano (Brasil)

```python
brasil = df_filtrado[df_filtrado['country'] == 'Brazil'].copy()
brasil['daily_deaths'] = brasil['total_deaths'].diff().fillna(0)
brasil['year'] = brasil['date'].dt.year
brasil['month'] = brasil['date'].dt.month

# Agrupamento por mês e ano
heatmap_data = brasil.groupby(['year', 'month'])['daily_deaths'].sum().unstack().fillna(0)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap='Reds', annot=True, fmt='.0f', linewidths=0.5)

plt.title('Heatmap de Mortes Mensais por COVID-19 no Brasil')
plt.xlabel('Mês')
plt.ylabel('Ano')
plt.tight_layout()

plt.savefig('../images/heatmap_mortes_mensais_brasil.png')

plt.show()
```
