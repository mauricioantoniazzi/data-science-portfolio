### Importação e carregamento

```python
import pandas as pd

# Carregar o CSV
df = pd.read_csv('../data/listings.csv')

# Visualizar primeiras linhas
df.head()
```

```python
df.describe(include='all')
```

### Limpeza e Seleção de Colunas Relevantes

```python
colunas_uteis = [
    'id', 'name', 'neighbourhood_cleansed', 'latitude', 'longitude',
    'room_type', 'property_type', 'price', 'minimum_nights', 'maximum_nights',
    'accommodates', 'bathrooms_text', 'bedrooms', 'beds', 'amenities',
    'availability_365', 'number_of_reviews', 'review_scores_rating',
    'review_scores_cleanliness', 'review_scores_location', 'instant_bookable'
]

df = df[colunas_uteis].copy()
```

```python
# converter price para numero
df['price'] = df['price'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
# Tratar valores ausentes
df.isnull().sum().sort_values(ascending=False)
```

```python
# Preencher notas com 0 onde não há reviews
df['review_scores_rating'] = df['review_scores_rating'].fillna(0)
df['review_scores_cleanliness'] = df['review_scores_cleanliness'].fillna(0)
df['review_scores_location'] = df['review_scores_location'].fillna(0)

# Preencher campos categóricos com 'Não informado'
df['bathrooms_text'] = df['bathrooms_text'].fillna('Não informado')
df['bedrooms'] = df['bedrooms'].fillna(0)
df['beds'] = df['beds'].fillna(0)
df['price'] = df['price'].fillna(0)
```

### Insight 1: Distribuição de Preços por Tipo de Quarto

#### Como variam os preços médios dos imóveis de acordo com o tipo de quarto (room_type)?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar preços absurdos (outliers)
df_preco = df[df['price'] < 1500]

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_preco, x='room_type', y='price', palette='Set3')

plt.title('Distribuição de Preços por Tipo de Quarto no Airbnb - Rio de Janeiro')
plt.xlabel('Tipo de Quarto')
plt.ylabel('Preço (R$)')
plt.xticks(rotation=45)
plt.tight_layout()

# plt.savefig('../images/precos_por_tipo_quarto.png')

plt.show()
```

### Insight 2: Quais bairros têm os anúncios mais caros?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Selecionar colunas relevantes
colunas_uteis = [
    'id', 'neighbourhood_cleansed', 'price'
]

df_filtrado = df[colunas_uteis].copy()

# Converter price para float
df_filtrado['price'] = df_filtrado['price'].replace(r'[\$,]', '', regex=True).astype(float)

# Filtrar valores extremos (preços acima de R$1500)
df_filtrado = df_filtrado[df_filtrado['price'] < 1500]

# Agrupar por bairro
media_bairros = df_filtrado.groupby('neighbourhood_cleansed').agg(
    media_preco=('price', 'mean'),
    total=('id', 'count')
).reset_index()

# Filtrar bairros com pelo menos 30 anúncios
media_bairros = media_bairros[media_bairros['total'] >= 30]

# Top 10 bairros com maior preço médio
top_bairros = media_bairros.sort_values(by='media_preco', ascending=False).head(10)

# Gerar gráfico
plt.figure(figsize=(12, 6))
sns.barplot(data=top_bairros, x='media_preco', y='neighbourhood_cleansed', hue='neighbourhood_cleansed', legend=False, palette='coolwarm')
plt.title('Top 10 Bairros com Maior Preço Médio - Airbnb RJ')
plt.xlabel('Preço Médio (R$)')
plt.ylabel('Bairro')
plt.tight_layout()

# Salvar imagem
# plt.savefig('../images/top_bairros_preco_medio.png')
plt.show()
```

### Insight 3: Quais regiões têm mais avaliações (popularidade)

```python
# Agrupamento por bairro com soma de avaliações
avaliacoes_bairros = df.groupby('neighbourhood_cleansed').agg(
    total_reviews=('number_of_reviews', 'sum'),
    total_anuncios=('id', 'count')
).reset_index()

# Top 10 bairros com mais avaliações acumuladas
top_bairros_reviews = avaliacoes_bairros.sort_values(by='total_reviews', ascending=False).head(10)

# Gráfico
plt.figure(figsize=(12, 6))
sns.barplot(data=top_bairros_reviews, x='total_reviews', y='neighbourhood_cleansed', hue='neighbourhood_cleansed', legend=False, palette='crest')
plt.title('Top 10 Bairros com Mais Avaliações no Airbnb - RJ')
plt.xlabel('Total de Avaliações')
plt.ylabel('Bairro')
plt.tight_layout()

plt.savefig('../images/top_bairros_reviews.png')

plt.show()
```

### Insight 4: Existe relação entre o preço do anúncio e a nota média recebida?

```python
# Filtrar dados relevantes
df_avaliacao = df[(df['price'] < 1500) & (df['review_scores_rating'] > 0)]

plt.figure(figsize=(10, 6))
sns.regplot(
    data=df_avaliacao,
    x='price',
    y='review_scores_rating',
    scatter_kws={'alpha':0.3},
    line_kws={'color': 'red'}
)
plt.title('Relação entre Preço e Nota Média - Airbnb RJ')
plt.xlabel('Preço da Diária (R$)')
plt.ylabel('Nota Média de Avaliação')
plt.tight_layout()

plt.savefig('../images/preco_vs_nota.png')

plt.show()
```

### Insight 5: Quais tipos de propriedade são mais comuns e como se comportam em relação a

```python
# Filtrar preços absurdos
df_filtrado = df[df['price'] < 1500]

# Agrupar por tipo de propriedade
tipos_propriedade = df_filtrado.groupby('property_type').agg(
    media_preco=('price', 'mean'),
    total_anuncios=('id', 'count')
).reset_index()

# Considerar apenas tipos com 30+ anúncios
tipos_filtrados = tipos_propriedade[tipos_propriedade['total_anuncios'] >= 30]

# Top 10 tipos mais comuns
top_tipos = tipos_filtrados.sort_values(by='total_anuncios', ascending=False).head(10)

# Gráfico
plt.figure(figsize=(12, 6))
sns.barplot(data=top_tipos, x='media_preco', y='property_type', hue='property_type', legend=False, palette='flare')
plt.title('Top 10 Tipos de Propriedade por Preço Médio - Airbnb RJ')
plt.xlabel('Preço Médio (R$)')
plt.ylabel('Tipo de Propriedade')
plt.tight_layout()

plt.savefig('../images/preco_por_tipo_propriedade.png')

plt.show()
```
